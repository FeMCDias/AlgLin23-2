```python
import matplotlib.pyplot as plt
from matplotlib import patches
xo = 3
yo = 3
xp = -1
yp = -1
fig = plt.figure(figsize=(8,5))
plt.plot([-1, -1], [-4, 4], c='r')
plt.plot([xo, xp], [yo, yp], c='b')
plt.plot([-2, 4], [0, 0], c='k')
plt.plot([0, 0], [-5, 5], c='k')
plt.plot([xo, xo], [0, yo], 'b:')
plt.scatter([xo], [yo], c='b')
plt.scatter([xp], [yp], c='r')
plt.scatter([0], [0], c='k')
plt.text(xo+0.1, yo+0.1, '($x_o$, $y_o$)')
plt.text(xo+0.1, yo+0.6, 'Objeto')
plt.text(xp+0.1, yp-0.1, '($x_p$, $y_p$)')
plt.text(xp+0.1, yp-0.6, 'Projeção')
plt.text(0+0.1, 0-0.4, '(0,0)')
plt.text(0+0.1, 0-1, 'Pinhole')
plt.text(-1.9, 4.5, "Câmera")
plt.xticks([])
plt.yticks([])
plt.grid('both')

# specify the location of (left,bottom),width,height
rect=patches.Rectangle((-2,-5),2,10, 
                        fill = True,
                        facecolor="red",
                        alpha=0.1,
                        linewidth = 2)
plt.gca().add_patch(rect)

plt.savefig('pinhole_diagrama.png')
```
